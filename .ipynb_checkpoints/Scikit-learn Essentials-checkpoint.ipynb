```python
import pandas as pd 
from sklearn.datasets import load_wine

wine_data = load_wine()
#convert data to pandas dataframe

wine_df =pd.DataFrame(wine_data.data,columns=wine_data.feature_names)
```

```python
#add the target label

wine_df["target"] = wine_data.target

#take a preview

wine_df.head()
```

```python
#in order to get some good results from machine learning algorithms we have to put some clean data
#if the data is bad the output will also be bad
#so do preprocessing
#now we preprocesses data to be sent to algorithm
#real world data contains- missing values, redundant values, outliers, errors and noise
#remove these or the data will incorporate these mistakes into its approximation functions
#so the machine will learn to make mistakes into its new instances
#garbedge in garbege out

#ML models requires - numaric data only

#to combat this problem - standardize the features using SKlearn's standard scaler class
#this class keeps the mean of feature is approximately equal to zero, so it will ensure the mean of each feature is 0
#Col name are the features
```

```python
from sklearn.preprocessing import StandardScaler

#Split data into features and label
X = wine_df[wine_data.feature_names].copy() #features
y = wine_df["target"].copy() #labels
```

```python
#next instantiate scaler and fit on features

scaler = StandardScaler() #features are stored here after scaling - here we created and fitted on the features x using the fit 
                            # method
scaler.fit(X)               #this calculates the mean and the standard deviation of each features in x

#after feeting we trandform
```

```python
#transform features of X into a scaled version
X_scaled = scaler.transform(X.values) #what this scaling does is substract the mean and dividing by the standard deviation of
                                        #each feature
                                    
#this code performs feature scaling on wine data set using standardscaler from sciket learn library
```

```python
#now the first instance of scale features is printed using the print function
print(X_scaled[0])
```

```python
print(X_scaled)
```

```python
#--------------------------X------------------------PreProcessing done-------------------------------------X-----------
```

```python

#Model Training and building the model

#before a ML model can make predictions it has to be trained on some set of data in order to learn an appoximation function

#how to know if the model performs well on data if it has not seen before - so we will not unless we test it out

#so now we will test a machine learning model before placing it in the real world enviroment is to split the date with training
#and testing

#once we split the data fromtrainign and testing, the test set will not have access to the train data.

#the ML model will have train data only 
#the model will not see the test data.

#scikit learn has builtin function to do so.

#keep learger portion for training
#70% for taining and 30% for model evaluation like so
#30%- to evaluate the model's ability to generalize to unseen instances
```

```python
#Model training

from sklearn.model_selection import train_test_split

#split data into train and test
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled,  #takes 2 arrays - X and Y, which represent the 
                                                                 y,           #the features and target variables resptively
                                                                 train_size=.7,
                                                                 random_state=25) #random_state is set to 25 to ensure
                                                                       #the same split is generated each time the code is run
                    #above function returns 4 arrays 
                                                                                        
#Check the split are correct   #f: formatted string literal
#print(f"Train size: {round(len(X_train_scaled)/len(X)*100)}% \n\ 
#Test size: {round(len(X_test_scaled)/len(X)*100)}%")



# more clean way to display the percantage:

# Check the splits are correct
train_size_percentage = round(len(X_train_scaled) / len(X) * 100)
test_size_percentage = round(len(X_test_scaled) / len(X) * 100)

print("Train size:", train_size_percentage, "%")
print("Test size:", test_size_percentage, "%")
```

```python
#we will build 3 machine learning models to predict the class of y
#one with logistic regression
#one with  support vector machine with svc(Support Vector Classifier) - best fit
#last one with decision tree classifier
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
# Instantiate the models
logistic_regression = LogisticRegression()
svm = SVC()
tree = DecisionTreeClassifier()
```

```python
#training the models:

logistic_regression.fit(X_train_scaled, y_train)
svm.fit(X_train_scaled, y_train)
tree.fit(X_train_scaled, y_train)
```

```python
#Now we test with the test data set
#making predictions with each model

log_reg_preds = logistic_regression.predict(X_test_scaled)
svm_preds = svm.predict(X_test_scaled)
tree_preds = tree.predict(X_test_scaled)

#we will look for accuracy - the number of correct instances predicted
```

```python
print (tree_preds)
```

```python
#Evaluation - final step to understand if the model is good or not
#it's evaluating the performance of 3 machine learning models

from sklearn.metrics import classification_report

#put the model predictions in a dictionary
#all data are stored in it. with key and value
model_preds = {
    "Logistic Regression": log_reg_preds,
    "Support Vector Machine": svm_preds,
    "Decision tree": tree_preds
}
#for loop is iterating through each model in the model_preds and print the classification report of each model values
#with the function classification_report(y_test,preds)
#y-tests: the true ones
#preds: the predicted ones


for model,preds in model_preds.items():
    print(f"{model} Results\n{classification_report(y_test,preds)}", sep="\n\n")
```

```python
#above report provides the summery  of precision recall and f1 score:F1-Score is a measure combining both precision and recall 
#for each class in the data set as well as the overall accuracy oand weighted averages
#the same parameter in the print function  is not necessary in this case as it is a default separated between arguments
```
